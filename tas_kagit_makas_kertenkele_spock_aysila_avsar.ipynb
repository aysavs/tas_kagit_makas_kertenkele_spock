```python
import random

def tas_kagit_makas_kertenkele_spock_aysila_avsar():
    """Taş, Kağıt, Makas, Kertenkele, Spock oyunu"""

    oyuncu_secenekleri = ["taş", "kağıt", "makas", "kertenkele", "spock"]

    # Oyun tanıtım ve kurallar
    print("Taş, Kağıt, Makas, Kertenkele, Spock Oyununa Hoş Geldiniz!\n")
    print("**Oyunun Amacı:**\n"
          "İlk iki turu kazanan oyunu kazanır!\n"
          "Seçenekleriniz: Taş, Kağıt, Makas, Kertenkele, Spock\n"
          "Her seçenek diğerlerini yener veya yenilir.\n"
          "Örneğin: Taş, makası ve kertenkeleyi yener.\n"
          "Ayrıntılı kurallar için oyun içindeki yardım menüsünü kullanabilirsiniz.\n")
    print("**Oyun Nasıl Oynanır:**\n"
          "1. Oyun size bir seçenek sunacaktır.\n"
          "2. Siz de karşılık gelen seçeneği yazıp enter tuşuna basın.\n"
          "3. Oyun, sizin ve bilgisayarın seçimlerini karşılaştırarak galibi belirleyecektir.\n"
          "4. İlk iki turu kazanan oyunun galibi olacaktır.\n"
          "5. Oyunu bırakmak için 'çıkış' yazabilirsiniz.\n")

    def bilgisayar_oyuna_devam_mi():
        """Bilgisayarın oyuna devam etmek isteyip istemediğini belirler."""
        return random.choice(["evet", "hayır"])

    def oyun_sonucu():
        oyuncu_galibiyet = 0
        bilgisayar_galibiyet = 0

        while oyuncu_galibiyet < 2 and bilgisayar_galibiyet < 2:
            # Oyuncunun seçimini al
            oyuncu_secimi = input("Seçiminizi yapın (taş, kağıt, makas, kertenkele, spock): ").lower()

            # 'Çıkış' komutu kontrolü
            if oyuncu_secimi == 'çıkış':
                print("Oyun sonlandırılıyor.")
                return None

            # Geçersiz giriş kontrolü
            while oyuncu_secimi not in oyuncu_secenekleri:
                print("Geçersiz seçim. Lütfen geçerli bir seçenek girin.")
                oyuncu_secimi = input("Seçiminizi yapın (taş, kağıt, makas, kertenkele, spock): ").lower()

                # 'Çıkış' komutu kontrolü
                if oyuncu_secimi == 'çıkış':
                    print("Oyun sonlandırılıyor.")
                    return None

            # Bilgisayarın seçimini rastgele belirle
            bilgisayar_secimi = random.choice(oyuncu_secenekleri)
            print(f"Bilgisayarın seçimi: {bilgisayar_secimi}")

            # Kazananı belirle ve galibiyet sayılarını güncelle
            if oyuncu_secimi == bilgisayar_secimi:
                print("Berabere!")
            elif (oyuncu_secimi == "taş" and bilgisayar_secimi in ["makas", "kertenkele"]) or \
                 (oyuncu_secimi == "kağıt" and bilgisayar_secimi in ["taş", "spock"]) or \
                 (oyuncu_secimi == "makas" and bilgisayar_secimi in ["kağıt", "kertenkele"]) or \
                 (oyuncu_secimi == "kertenkele" and bilgisayar_secimi in ["kağıt", "spock"]) or \
                 (oyuncu_secimi == "spock" and bilgisayar_secimi in ["taş", "makas"]):
                print("Tebrikler, kazandınız!")
                oyuncu_galibiyet += 1
            else:
                print("Üzgünüm, kaybettiniz.")
                bilgisayar_galibiyet += 1

            # Turun sonucunu ekrana yazdır
            print(f"Tur sonucu: Oyuncu {oyuncu_secimi}, Bilgisayar {bilgisayar_secimi}")

        # Oyunun galibini belirle
        if oyuncu_galibiyet == 2:
            print("Tebrikler! Oyunu kazandınız.")
            return True
        elif bilgisayar_galibiyet == 2:
            print("Üzgünüm, oyunu kaybettiniz.")
            return False

    toplam_oyun_sayisi = 0
    oyuncu_toplam_galibiyet = 0
    bilgisayar_toplam_galibiyet = 0

    while True:
        toplam_oyun_sayisi += 1
        galip_mi = oyun_sonucu()
        if galip_mi is None:
            break  # Oyun 'çıkış' komutu ile sonlandırıldı
        elif galip_mi:
            oyuncu_toplam_galibiyet += 1
        else:
            bilgisayar_toplam_galibiyet += 1

        # Genel istatistikleri göster
        print(f"Toplam oyun sayısı: {toplam_oyun_sayisi}")
        print(f"Oyuncu toplam galibiyet: {oyuncu_toplam_galibiyet}")
        print(f"Bilgisayar toplam galibiyet: {bilgisayar_toplam_galibiyet}")

        # Yeni bir oyun oynamak isteyip istemediklerini sorma
        devam_etmek = input("Yeni bir oyun oynamak ister misiniz? (evet/hayır): ").lower()
        if devam_etmek == "çıkış" or devam_etmek == "hayır":
            print("Oyun sonlandırılıyor.")
            break

        # Bilgisayarın yeni oyun isteyip istemediğini belirleme
        bilgisayar_devam_etmek = bilgisayar_oyuna_devam_mi()
        if devam_etmek == "evet" and bilgisayar_devam_etmek == "evet":
            print("Bilgisayar da yeni oyuna ever diyor! Yeni oyun başlatılıyor...")
        else:
            print("Bir taraf oyuna devam etmek istemiyor. Oyun sonlandırılıyor.")
            break

# Oyunu başlat
tas_kagit_makas_kertenkele_spock_aysila_avsar()

```
